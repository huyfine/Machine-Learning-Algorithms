### 2.1 Normalize data

```python
def dataset_minmax(dataset):
    minmax = list()
    for i in range(len(dataset[0])):
        col_values = [row[i] for row in dataset]
        value_min = min(col_values)
        value_max = max(col_values)
        minmax.append([value_min, value_max])
    return minmax

dataset = [[50,30], [20,90]]
print(dataset)
minmax = dataset_minmax(dataset)
print(minmax)
```

$\text{scaled value} = \frac{\text{value - min}}{\text{max - min}}$

```python
def normalize_dataset(dataset, minmax):
    for row in dataset:
        for i in range(len(row)):
            row[i] = (row[i] - minmax[i][0]) / (minmax[i][1] - minmax[i][0])
```

```python
dataset = [[50,30],[20,90]]
minmax = dataset_minmax(dataset)
normalize_dataset(dataset,minmax)
print(dataset)
```

### 2.2 Standarlize data

```python
def column_means(dataset):
    means = [0 for i in range(len(dataset[0]))]
    for i in range(len(dataset[0])):
        col_values = [row[i] for row in dataset]
        means[i] = sum(col_values) / float(len(dataset))

    return means
```

$ \text{standard deviation} = \sqrt{\frac{\sum_{i=1}^{n} (\text{value}_i - \text{mean})^2}{\text{count}(\text{values}) - 1}} $

```python
def column_stdevs(dataset, mean):
    stdevs = [0 for i in range(len(dataset[0]))]
    for i in range(len(dataset[0])):
        variance = [pow(row[i] - means[i], 2) for row in dataset]
        stdevs[i] = sum(variance)
    stdevs = [sqrt(x / (float(len(dataset) - 1))) for x in stdevs]
    return stdevs
```

```python
def standardlize_dataset(dataset, means, stdevs):
    for row in dataset:
        for i in range(len(row)):
            row[i] = (row[i] - means[i]) / stdevs[i]

standardlize_dataset(dataset, means, stdevs)
print(dataset)
```
